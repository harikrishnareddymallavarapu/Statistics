```
import numpy as np
from numpy.random import randint
import seaborn as sns
import statistics as stat
```

```
#A Lift manufacturer wants to find the probability that a lift will snap
# Conditions - 20 persons 1700 kgs 
```

```
randint(60,100,20).sum()
```

```
#let me take thousand samples and check if atleast it fails
def runMultipleSamples(noOfIterations):
  sampleSums = []
  for i in range(noOfIterations):
    sampleSums.append(randint(60,100,20).mean())
  return sampleSums
```

```
sampleData = runMultipleSamples(300)
```

```
sns.distplot(sampleData)
```

```
from scipy.stats import skewnorm
```

```
s = skewnorm.rvs(12, size=10000)
```

```
sns.distplot(s)
```

```
def sample_mean_calculator(population_array, sample_size, n_samples):
  sample_means = []
  for i in range(n_samples):
    sample = np.random.choice(population_array, size=sample_size, replace=False)
    sample_mean = stat.mean(sample)
    sample_means.append(sample_mean)
  return sample_means
```

```
_ = sns.distplot(sample_mean_calculator(s, 10, 1000))
```
